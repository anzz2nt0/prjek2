```python
import matplotlib.pyplot as plt
import numpy as np

months = np.arange(1, 13)
rainfall_1981_1990 = [400, 350, 300, 250, 200, 150, 100, 50, 100, 150, 200, 250]
rainfall_1991_2000 = [450, 400, 350, 300, 250, 200, 150, 100, 150, 200, 250, 300]
rainfall_2001_2010 = [500, 450, 400, 350, 300, 250, 200, 150, 200, 250, 300, 350]

bar_width = 0.25

r1 = np.arange(len(rainfall_1981_1990))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

plt.bar(r1, rainfall_1981_1990, color='blue', width=bar_width, edgecolor='grey', label='1981 - 1990')
plt.bar(r2, rainfall_1991_2000, color='red', width=bar_width, edgecolor='grey', label='1991 - 2000')
plt.bar(r3, rainfall_2001_2010, color='green', width=bar_width, edgecolor='grey', label='2001 - 2010')

plt.xlabel('Bulan', fontweight='bold')
plt.ylabel('Curah Hujan (mm)', fontweight='bold')
plt.xticks([r + bar_width for r in range(len(rainfall_1981_1990))], months)

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

time = ["00.00 UTC", "03.00 UTC", "06.00 UTC", "09.00 UTC", "12.00 UTC", "15.00 UTC", "18.00 UTC", "21.00 UTC", "00.00 UTC"]
asimilasi = [23.5, 25.0, 26.0, 25.5, 24.0, 23.0, 22.5, 23.0, 24.5]
observasi = [24.0, 25.5, 26.5, 26.0, 25.0, 24.5, 23.5, 24.0, 25.5]
non_asimilasi = [23.0, 24.5, 25.5, 24.5, 23.5, 22.0, 21.5, 22.0, 23.0]

plt.figure(figsize=(10, 6))
plt.plot(time, asimilasi, marker='o', color='blue', label='Asimilasi')
plt.plot(time, observasi, marker='o', color='red', label='Observasi')
plt.plot(time, non_asimilasi, marker='o', color='orange', label='Non Asimilasi')

plt.xlabel('Waktu', fontweight='bold')
plt.ylabel('Celcius', fontweight='bold')
plt.title('Suhu Permukaan')
plt.xticks(rotation=45)
plt.legend()

plt.grid(True)
plt.show()
```
